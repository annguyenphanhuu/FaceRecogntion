##Result

```python
import cv2
import os
from google.colab.patches import cv2_imshow

# Đường dẫn đến thư mục chứa ảnh
folder_path = '/content/drive/MyDrive/Project/Biometric/FaceRecog/src/TestImage/After'

# Lấy danh sách tất cả các tệp ảnh trong thư mục
image_files = [f for f in os.listdir(folder_path) if f.endswith(('.jpg', '.jpeg', '.png', '.bmp'))]

# Tỉ lệ co giãn theo chiều rộng và chiều cao (ví dụ: co giãn về 50%)
scale_percent = 75  # Tỉ lệ co giãn

# Hiển thị từng ảnh một
for image_file in image_files:
    # Đường dẫn đầy đủ đến tệp ảnh
    image_path = os.path.join(folder_path, image_file)

    # Đọc ảnh từ tệp
    image = cv2.imread(image_path)

    # Kiểm tra xem ảnh có được đọc thành công không
    if image is not None:
        # Tính toán tỉ lệ co giãn cho chiều rộng và chiều cao
        width = int(image.shape[1] * scale_percent / 100)
        height = int(image.shape[0] * scale_percent / 100)

        # Co giãn ảnh theo tỉ lệ đã tính toán
        resized_image = cv2.resize(image, (width, height))

        # Hiển thị ảnh đã co giãn
        cv2_imshow(resized_image)
    else:
        print(f"Không thể đọc được ảnh từ tệp {image_file}")
```
